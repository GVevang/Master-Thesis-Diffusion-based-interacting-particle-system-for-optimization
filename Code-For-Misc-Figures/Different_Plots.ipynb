### 3D - Plots

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D 
 
X = np.linspace(-3, 3, 100)     
Y = np.linspace(-3, 3, 100)     
X, Y = np.meshgrid(X, Y) 
Z = (X**2 - 10 * np.cos(2 * np.pi * X)) + \
  (Y**2 - 10 * np.cos(2 * np.pi * Y)) + 20

fig = plt.figure() 
ax = fig.add_subplot(111, projection='3d')  # Use add_subplot for 3D plots
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
  cmap=cm.nipy_spectral, linewidth=0.08,
  antialiased=True)


ax.view_init(elev=35, azim=35)

ax.plot([0, 0], [0, 0], [0, 100], color='blue', linestyle='-',zorder=10)

# Adding text for the global minimum
ax.text(0, 0, 100, "Global Minimum at [0, 0]", color='blue')
plt.savefig('Rast_3D.eps', format='eps')    
plt.show()
```
