```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
from matplotlib.colors import Normalize

def Rast(X,A):
    test = A*X+np.sum(np.square(X)-A*np.cos(2*np.pi*X),axis=0)
    #test = A+np.sum(np.square(X)-A*np.cos(2*np.pi*X),axis=0)/X.shape[0]
    return test    
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
ite = 1000
A = 10
beta_lin = np.zeros([ite])
x_lin = np.zeros([ite])
x_lin = np.linspace(-2.2,2.2,10000)
alpha = np.array([1,5,10])
for x in range(3):
    beta_lin = np.exp(-alpha[x]*(A*1+np.square(x_lin)-A*np.cos(2*np.pi*x_lin)))
    plt.plot(x_lin,beta_lin,label=r'$\beta$ = ' + str(alpha[x]))
#beta_lin = (A*1+np.square(x_lin)-A*np.cos(2*np.pi*x_lin))

# Plot
plt.plot(x_lin,beta_lin)
#plt.title(r'$\beta$ vs. Iteration')
plt.xlabel('X',fontsize=24)
plt.ylabel('PDF',fontsize=24)
plt.tick_params(axis='x', labelsize=15)
plt.tick_params(axis='y', labelsize=15)
plt.legend()
plt.grid(True)
plt.savefig('distribution2.eps', format='eps')    
plt.show()
```
